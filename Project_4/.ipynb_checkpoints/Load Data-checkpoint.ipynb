```python
import pandas as pd
import pickle
```

```python
file = open("reviews2.obj",'rb')
reviews = pickle.load(file)
file.close()

new_reviews = []

for review in reviews:
    if(review):
        new_reviews.append(review)

reviews = new_reviews
```

```python
#Statistics
print('Reviews: {}'.format(len(reviews)))
count = 0
for review in reviews:
    if review['posts']:
        count += len(review['posts'])

print('Posts: {}'.format(count))

count = 0
for review in reviews:
    if review['worthit']:
        count += 1
print('Worth It/Not Worth It: {}'.format(count))

count = 0
for review in reviews:
    if review['price']:
        count += 1
print('Price: {}'.format(count))

count = 0
for review in reviews:
    if review['doctor']:
        count += 1
print('Doctor: {}'.format(count))

count = 0
for review in reviews:
    if review['location']:
        count += 1
print('Location It: {}'.format(count))

```
